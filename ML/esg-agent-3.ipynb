```python
import pandas as pd
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("rahuljangir78/esg-and-financial-metrics-of-manufacturing-firms")

print("Path to dataset files:", path)
```

```python
df = pd.read_csv("/kaggle/input/datasets/rahuljangir78/esg-and-financial-metrics-of-manufacturing-firms/Manufacturing_ESG_Financial_Data.csv")
```

```python
df = df[["E_Score","S_Score","G_Score","Board_Independence","Industry_Type"]]
```

```python
def check_social(row):
    return "Violation" if row["S_Score"] < 60 else "Compliant"

def check_governance(row):
    if row["G_Score"] < 60 or row["Board_Independence"] < 0.5:
        return "Violation"
    return "Compliant"
def check_environment(row):
    return "Violation" if row["E_Score"] < 60 else "Compliant"
```

```python
df["E_Compliance"] = df.apply(check_environment, axis=1)
df["S_Compliance"] = df.apply(check_social, axis=1)
df["G_Compliance"] = df.apply(check_governance, axis=1)
```

```python
def overall_status(row):
    if "Violation" in [row["E_Compliance"],
                       row["S_Compliance"],
                       row["G_Compliance"]]:
        return "Non-Compliant"
    return "Compliant"

df["Overall_Compliance"] = df.apply(overall_status, axis=1)
```

```python
AGENT2_output = df[[
    "E_Compliance",
    "S_Compliance",
    "G_Compliance",
    "Overall_Compliance",
    
]]
```

```python
AGENT2_output.to_csv("ESG1_Output.csv", index=False)
```
